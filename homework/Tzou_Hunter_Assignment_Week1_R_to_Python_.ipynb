## Part A -- Translation (Conceptual)

For the R code below:

1.  Predict the output\
2.  Translate it into Python\
3.  Compute the result in Python

``` r
x <- c(1, 3, 5, 7)

mean(y)
```

Clearly show your Python code and final result.

------------------------------------------------------------------------

````python
## I predict that the code will check if the element of the vector stored in x is greater than 4. If it is, than square it. If it is not, than add 1 to that element instead. Then take the mean of the new array which is stored in y.
## I predict the new array to be [2, 4, 25, 49] and the mean to be 20.

## Python Code
import numpy as np

x = np.array([1, 3, 5, 7])
y = np.where(x>4, x**2,x+1)

print(y)
````
