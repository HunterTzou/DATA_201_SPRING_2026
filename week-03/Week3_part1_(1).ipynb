## 0. Setup

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
````

## 1. Data for Visualization

Simulated housing-style dataset.

````python
np.random.seed(42)

df = pd.DataFrame({
    "price": np.random.normal(250, 40, 100),
    "size": np.random.normal(1600, 200, 100),
    "bedrooms": np.random.choice([2, 3, 4], size=100),
    "neighborhood": np.random.choice(["A", "B", "C"], size=100)
})

df.head()
````

## 2. First Plot: Scatterplot

In ggplot2:
```r
ggplot(df, aes(size, price)) + geom_point()
```

**TODO:**  
1. Create a **scatterplot** of `size` vs `price`  
2. **Label** the x-axis and y-axis

````python
sns.scatterplot(data=df, x="size", y="price")
plt.xlabel("Size")
plt.ylabel("Price")
````

## 3. Adding Color (Aesthetics)

In ggplot2:
```r
ggplot(df, aes(size, price, color = neighborhood)) + geom_point()
```

**TODO:** Recreate the plot above using **seaborn**.

**Question:** What does seaborn do automatically that matplotlib does not?

````python
sns.scatterplot(data=df, x="size", y="price",hue="neighborhood")
plt.xlabel("Size")
plt.ylabel("Price")
````

## 4. Faceting vs Small Multiples

In ggplot2:
```r
ggplot(df, aes(size, price)) + geom_point() + facet_wrap(~ bedrooms)
```

**TODO:** Create **small multiples** by `bedrooms`.

````python
small_multiples = sns.FacetGrid(df, col="bedrooms")

small_multiples.map_dataframe(
    sns.scatterplot,
    x="size",
    y="price",
    hue="neighborhood"
)

small_multiples.set_axis_labels("Size", "Price")
small_multiples.set_titles("{col_name} Bedrooms")
small_multiples.add_legend()
````
